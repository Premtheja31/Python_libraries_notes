```python
fig,a=plt.subplots()
africa=a.plot(years,sales_Africa)[0]
africa.set_label("AFrica")
asp, = a.plot(years, sales_Asia_Pacific)
asp.set_label("Asia Pacific")
eur, = a.plot(years, sales_Europe)
eur.set_label("Europe")
eur.set_dashes([2,10,78,4])
plt.legend()#used to get the names and marked lines as given below
plt.show()
```
