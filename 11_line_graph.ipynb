```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

line graph is used to present the trend with time variable on the x-axis
in some cases it is used as alternative to scatter plot yo understand the relation between 2 variables

```python
months = np.array(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
sales = np.array([241268.56, 184837.36, 263100.77, 242771.86, 288401.05, 401814.06, 258705.68, 456619.94, 481157.24, 422766.63, 555279.03, 503143.69])
```

```python
plt.plot(months,sales)
plt.xticks(rotation=90)
plt.show()
```

```python
months = np.array(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
sales = np.array([241268.56, 184837.36, 263100.77, 242771.86, 288401.05, 401814.06, 258705.68, 456619.94, 481157.24, 422766.63, 555279.03, 503143.69])
sales2 = np.array([261268.56, 194837.36, 283100.77, 292771.86, 228401.05, 421814.06, 278705.68, 486619.94, 451157.24, 442766.63, 575279.03, 593143.69])
plt.xticks(rotation=90)
plt.plot(months,sales,"red",sales2,"blue")
plt.show()
plt.xticks(rotation=90)
plt.plot(months,sales,"red",marker=">")
plt.show()
plt.xticks(rotation=90)

plt.plot(months,sales2,"blue",marker="+")
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.plot(months,sales,"red",marker=">")
plt.xticks(rotation=90)

plt.plot(months,sales2,"blue",marker="+")
plt.show()
```
