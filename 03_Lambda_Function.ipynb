```python
import pandas as pd
import numpy as np
```

```python
values= np.array([[60,78,65,52,85],
         [75,85,62,61,60],
         [89,76,56,51,35],
         [40,60,70,40,45],
         [50,70,75,46,85],
         [45,66,77,49,78]])
```

```python
marks_df=pd.DataFrame(values,columns=["english","physics","chemistry","maths","biology"],index=["prem","varshith","sanjeev","sandesh","vinay","hemanth"])
```
